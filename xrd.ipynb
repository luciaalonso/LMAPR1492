```python
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter
from pymatgen.analysis.diffraction.xrd import XRDCalculator
from pymatgen.analysis.diffraction.core import AbstractDiffractionPatternCalculator
from matplotlib import pyplot
from IPython.display import Markdown,display,Image

def printmd(string):
    display(Markdown(string))

with MPRester("Niub9m7cC3zFHMd6") as m:
    structure = m.get_structure_by_material_id("mp-36111")
    w = CifWriter(structure)
    w.write_file('mystructure.cif')

    
```

```python
print('Quand on envoie des rayons X sur des cristaux, on observe differents phénomènes. Ces phénomènes sont des interferences,lesquels peuvent être destructives ou constructives (en dépendant de la direction). A partir de cette image de diffraction, il est possible, sous certaines conditions, de déterminer la position de tous les atomes de la structure. ')
#Création de l'objet
a = XRDCalculator()
#Calcul des pics de diffraction
b = a.get_pattern(structure)
a.get_plot(structure)
pyplot.show()
```

```python
print('A partir de l image de diffraction, il est possible, sous certaines réserves de déterminer la position de tous les atomes de la structure. Dans cette partie, nous allons nous intéresser aux pics du diffractogramme pour un rayonnement X incident CuK\u03B1 de longueur d’onde \u03BB=1.54060 Angstrom sur le matériau ScCuO\u2082 ')
print('Comme demandé, nous allons principalement nous concentrer sur les trois premiers pics ')

printmd ('**PREMIER PIC DU DIFFRACTOGRAMME**') 

#calcul du premier pic
i1 = b.y[0] #intensité
angle1 = b.x[0] #2theta
hkl1 = b.hkls[0] #indices de Miller

print('intensité :',i1) 
print('2\u03B8 :',angle1)
print('hkl :',hkl1)


printmd ('**DEUXIÈME PIC DU DIFFRACTOGRAMME**')

#calcul du deuxième pic
i2nég = b.y[1] #intensité
angle2nég = b.x[1] #2theta
hkl2nég = b.hkls[1] #indices de Miller

print('intensité :',i2nég) 
print('2\u03B8 :',angle2nég)
print('hkl :',hkl2nég)

print('On va négliger ce deuxième pic car il n est pas trop visible dans le diffractogramme et on voit que l intensité est trop')
print('faible. On considère donc le pic suivant.')
print('\n')

i2 = b.y[2] #intensité
angle2 = b.x[2] #2theta
hkl2 = b.hkls[2] #indices de Miller

print('intensité :',i2) 
print('2\u03B8 :',angle2)
print('hkl :',hkl2)

printmd ('**TROISIÈME PIC DU DIFFRACTOGRAMME**')

#calcul du troisième pic
i3 = b.y[3] #intensité
angle3 = b.x[3] #2theta
hkl3 = b.hkls[3] #indices de Miller

print('intensité :',i3) 
print('2\u03B8 :',angle3)
print('hkl :',hkl3)

printmd ('**Les trois premiers pics du diffractogramme pour un rayonnement X incident CuKα pour le matériau LiMgP**')

xaxis = [angle1,angle2,angle3]
yaxis = [i1,i2,i3]
width = 0.25
pyplot.bar(xaxis,yaxis,width,color=(0.3,0.5,0.7,1.0))
pyplot.xlabel('2\u03B8')
pyplot.ylabel('Intensité')
pyplot.title('Diffractogramme des trois premiers pics')
pyplot.show()
```
